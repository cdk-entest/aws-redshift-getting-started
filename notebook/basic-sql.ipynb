## Join Tables

```redshift
select customer.c_name, sum(orders.o_totalprice) as totalprice  
from testshare.public.customer, testshare.public.orders 
where testshare.public.customer.c_custkey = testshare.public.orders.o_custkey
group by customer.c_name 
order by totalprice desc 
limit 1000;
```

```redshift
select customer.c_name, sum(orders.o_totalprice) as totalprice  
from testshare.public.customer join testshare.public.orders on testshare.public.customer.c_custkey = testshare.public.orders.o_custkey
group by customer.c_name 
order by totalprice desc 
limit 1000;
```

## Unload Data to S3 

- Check how 128 RPU parallel move data to S3 in partitions 
- [docs unload](https://docs.aws.amazon.com/redshift/latest/dg/r_UNLOAD.html)

```redshift
-- PARALLEL ON FORMAT AS PARQUET DELIMITER 
UNLOAD ('select * from customer where customer.c_custkey in
(select customer.c_custkey from customer limit 1000)')
TO 's3://data-lake-xxx/redshift-parquet/'
IAM_ROLE default
PARALLEL ON FORMAT AS PARQUET;
```

```redshift
-- PARALLEL OFF FORMAT AS PARQUET | CSV | JSON 
UNLOAD ('select * from customer where customer.c_custkey in
(select customer.c_custkey from customer limit 1000)')
TO 's3://data-lake-xxx/redshift-parquet/'
IAM_ROLE default
PARALLEL OFF FORMAT AS PARQUET;
```

```redshift
-- PARALLEL OFF FORMAT AS CSV DELIMITER 
UNLOAD ('select * from customer where customer.c_custkey in
(select customer.c_custkey from customer limit 1000)')
TO 's3://data-lake-xxx/redshift-csv/'
IAM_ROLE default
PARALLEL OFF FORMAT AS CSV DELIMITER '|';
```
